```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
```

```python
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
```

```python
def create_dataset_pipeline(x, y, batch_size=32, training=True):
    """Most memory-efficient approach - resize during training"""
    dataset = tf.data.Dataset.from_tensor_slices((x, y))
    
    if training:
        dataset = dataset.shuffle(1000)
    
    def resize(image, label):
        # Resize on GPU during training
        image = tf.image.resize(image, [224, 224])
        return image, label
    
    dataset = dataset.map(resize, num_parallel_calls=tf.data.AUTOTUNE)
    dataset = dataset.batch(batch_size)
    dataset = dataset.prefetch(tf.data.AUTOTUNE)
    
    return dataset

batch_size = 16  # Small batch to save RAM

train_ds = create_dataset_pipeline(x_train, y_train, batch_size, training=True)
test_ds = create_dataset_pipeline(x_test, y_test, batch_size, training=False)
```

```python
print(train_ds)
```

```python
def visualize_batch(images_batch, labels_batch, num_samples=9):
    """
    Visualize images and labels from a batch
    """
    fig, axes = plt.subplots(3, 3, figsize=(12, 12))
    axes = axes.ravel()
    
    for i in range(min(num_samples, len(images_batch))):
        # Display image
        axes[i].imshow(images_batch[i])
        axes[i].set_title(f'Label: {labels_batch[i][0]}', fontsize=12)
        axes[i].axis('off')
    
    # Hide unused subplots
    for i in range(num_samples, len(axes)):
        axes[i].axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
for images, labels in train_ds.take(1):
    visualize_batch(images.numpy(), labels.numpy())
```

```python
for images, labels in test_ds.take(1):
    visualize_batch(images.numpy(), labels.numpy())
```

```python
pretrained_base = tf.keras.applications.MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet' # load weights trained on ImageNet
)
```

```python
pretrained_base.summary()
```

```python
pretrained_base.trainable = False # freeze base model
```

```python
pretrained_base.summary()
```

```python
transfer_model = models.Sequential([
    pretrained_base,
    layers.GlobalAveragePooling2D(name='gap_tl'),
    layers.Dense(128, activation='relu', name='fc_tl'),
    layers.Dropout(0.5, name='dropout_tl'),
    layers.Dense(10, activation='softmax', name='output_tl')
])
```

```python
transfer_model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
transfer_model.summary()
```

```python
transfer_history = transfer_model.fit(
    train_ds,
    epochs=10,
    validation_data=test_ds,
        verbose=1
)
```

```python
def loss_acc_plots(history):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='train_loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    plt.title('Loss')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='train_acc')
    plt.plot(history.history['val_accuracy'], label='val_acc')
    plt.title('Accuracy')
    plt.legend()
    plt.show()
```

```python
loss_acc_plots(transfer_history)
```

```python
loss, accuracy = transfer_model.evaluate(test_ds, verbose=0)
print(f"Test Loss: {loss}")
print(f"Test Accuracy: {accuracy}")
```

Loss Plot: Training loss steadily decreases, indicating improving fit to training data. Validation loss is lower than training loss and fluctuates slightly, suggesting decent generalization with some variance.

Accuracy Plot: Training accuracy steadily increases, approaching about 80%. Validation accuracy remains higher than training accuracy and is stable around 78-81%, indicating good generalization.

Overall, the model is learning well without strong overfitting since validation metrics are stable or improving alongside training. The consistent validation accuracy above training accuracy can happen due to regularization or batch effects.
